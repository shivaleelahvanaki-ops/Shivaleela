```
# ---------------------------------------------
# Simple Rule-Based Chatbot
# ---------------------------------------------

print("=" * 50)
print("🤖 Welcome to Simple NLP Chatbot")
print("Type 'bye' to exit.")
print("=" * 50)

while True:
    user = input("\nYou: ").lower()

    if user == "hello" or user == "hi":
        print("Bot: Hello! How can I help you?")

    elif "how are you" in user:
        print("Bot: I am doing great. Thank you for asking!")

    elif "your name" in user:
        print("Bot: My name is AI Chatbot.")

    elif "college" in user:
        print("Bot: I am designed to assist students in learning NLP.")

    elif "python" in user:
        print("Bot: Python is one of the most popular programming languages for AI and NLP.")

    elif "nlp" in user:
        print("Bot: NLP stands for Natural Language Processing.")

    elif "ai" in user:
        print("Bot: AI stands for Artificial Intelligence.")

    elif "thank" in user:
        print("Bot: You're welcome!")

    elif user == "bye":
        print("Bot: Goodbye! Have a nice day.")
        break

    else:
        print("Bot: Sorry, I don't understand that. Please try another question.")
```

```
from deep_translator import GoogleTranslator

# Input text
text = "The students are studying Artificial Intelligence in the university."

# Dictionary of languages
languages = {
    "Hindi": "hi",
    "Kannada": "kn",
    "Telugu": "te",
    "Tamil": "ta",
    "Malayalam": "ml",
    "French": "fr",
    "German": "de",
    "Spanish": "es",
    "Japanese": "ja",
    "Bengali": "bn"
}

print("Original Text:")
print(text)
print("\nTranslations:\n")

# Translate into each language
for language, code in languages.items():
    translated = GoogleTranslator(source='auto', target=code).translate(text)
    print(f"{language}:")
    print(translated)
    print("-" * 50)
```
